```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline

from iccpy.gadget import load_snapshot
from iccpy.gadget.labels import cecilia_labels
```

```python

number_of_snapshots = 200   # actually includes snap_000

file = '/home/martin/Documents/Tesis/outputs/snap_{:03d}'

snap = load_snapshot(file.format(200), label_table = cecilia_labels)

gas_pos = snap['POS '][0]

gas_x = gas_pos[:,0]
gas_y = gas_pos[:,1]
gas_z = gas_pos[:,2]

gas_r = np.sqrt(gas_x**2 + gas_y**2 + gas_z**2)

gas_U = snap['U   '][0]

gas_rho = snap['RHO '][0]
```

```python
plt.figure(figsize= (11,8))

_ = plt.hist(np.log(gas_U), bins = 500)
```

```python
plt.figure(figsize=(9,8))
ax = plt.gca()

plt.scatter(gas_r, gas_U, c = np.log(gas_rho), s = 0.05, cmap = 'coolwarm')
plt.xlabel('$r$', fontsize = 15)
plt.ylabel('$\\mathcal{U}$', fontsize = 15)
ax.set_xscale('log')
ax.set_yscale('log')
cb = plt.colorbar()
cb.set_label('log $\\rho$', fontsize = 15, rotation = 270)
plt.grid(True)
```

```python
plt.figure(figsize=(9,8))
ax = plt.gca()

plt.scatter(gas_x, gas_y, c = np.log(gas_U), s = 0.05, cmap = 'plasma')
plt.xlim([-500, 500])
plt.ylim([-500, 500])
plt.xlabel('$x$', fontsize = 15)
plt.ylabel('$y$', fontsize = 15)
cb = plt.colorbar()
cb.set_label('log $\\mathcal{U}$', fontsize = 15, rotation = 270)
plt.grid(True)
```

```python
fig = plt.figure(figsize=(9,8)) 
ax3 = Axes3D(fig)

ax3.set_xlim3d(-300, 300)
ax3.set_ylim3d(-300, 300)
ax3.set_zlim3d(-300, 300)

ax3.scatter(gas_x, gas_y, gas_z, c = np.log(gas_U), s = 0.1, cmap = 'plasma')
```
